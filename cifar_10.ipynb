```python
from src.models import Sequential
from src.layers import *
from src.activation_functions import ReLU
from src.activation_functions import Softmax
from src.loss_functions import CrossEntropy

import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import OneHotEncoder

from sklearn.utils import shuffle as sk_shuffle
```

# Load data

```python
from tensorflow.keras.datasets.cifar10 import load_data as load_data_CIFAR

(x_train, y_train), (x_test, y_test) = load_data_CIFAR() # MNIST

x_train = x_train.astype(np.float32)
x_test = x_test.astype(np.float32)

x_train /= 255
x_test /= 255
```

# Optional shuffle

```python
x_train, y_train = sk_shuffle(x_train, y_train)
x_test, y_test = sk_shuffle(x_test, y_test)
```

# Prepare data

```python
x_train_expanded = np.einsum('bhwc -> bchw', x_train)
x_test_expanded = np.einsum('bhwc -> bchw', x_test)
```

```python
one_hot_encoder = OneHotEncoder(sparse_output=False)

y_train_encoded = one_hot_encoder.fit_transform(y_train.reshape(-1, 1))
y_test_encoded = one_hot_encoder.fit_transform(y_test.reshape(-1, 1))
```

# Check sample

```python
plt.imshow(np.einsum("chw -> hwc", x_train_expanded[0]))
```

# Create model

```python
model = Sequential(
    input_size=(3, 32, 32),
    output_size=10
)
```

```python
model.add(Conv2D(
    n_outputs=3,
    filter_size=5,
    stride=2,
    padding=2
))
model.add(Activation(ReLU()))
model.add(Conv2D(
    n_outputs=3,
    filter_size=3,
    stride=1,
    padding=2
))
model.add(Activation(ReLU()))
model.add(Flatten())
model.add(Linear(n_outputs=120))
model.add(Activation(ReLU()))
model.add(Linear(n_outputs=84))
model.add(Activation(ReLU()))
model.add(Linear(n_outputs=10))
model.add(Activation(Softmax()))
```

```python
model.summary()
```

# Train model

```python
model.fit(
    training_data=[x_train_expanded[:10000], y_train_encoded[:10000]],
    epochs=20,
    learning_rate=0.0001,
    loss_function=CrossEntropy(),
    batch_size=100,
    verbose=True,
    validation_data=[x_test_expanded[:1000], y_test_encoded[:1000]]
)
```

# Evaluate model

```python
model.plot_history(['loss', 'accuracy'])
```

```python
y_pred = model.predict(x_test_expanded)
y_pred = np.argmax(y_pred, axis=1)
```

```python
cifar10_labels = [
    "airplane",
    "automobile",
    "bird",
    "cat",
    "deer",
    "dog",
    "frog",
    "horse",
    "ship",
    "truck"
]
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import f1_score, balanced_accuracy_score, accuracy_score

cm_model = confusion_matrix(y_test, y_pred)
ConfusionMatrixDisplay(cm_model, display_labels=cifar10_labels).plot(cmap='gray', xticks_rotation='vertical')

print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 score:", f1_score(y_test, y_pred, average='weighted'))
print("BA score:", balanced_accuracy_score(y_test, y_pred))
```
